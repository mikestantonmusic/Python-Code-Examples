```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import plotly.express as px
import numpy as np
```

```python
table=pd.read_html('https://www.basketball-reference.com/leagues/NBA_2020_per_game.html')
df=table[0]
df.head(10)
```

```python
#scatter plot

data_types = {'AST':'float','FGA':'float','Age':'float','PTS':'float'}


df=df[df.Age != "Age"]
df=df.loc[0:200,:]
df=df.astype(data_types)
```

```python


fig = px.scatter(df, x='FGA',y='AST',title='field goal average vs assists', width=1200, height=800)
fig.update_xaxes(title='FGA')
fig.update_yaxes(title='AST')
fig.show()
```

```python
#bar chart 

df=df[df.Age != "Age"]
df_top10=df.loc[0:10,:]
df_top10=df_top10.astype({'PTS':'float'})
df_top10=df_top10.astype({'FGA':'string'})

fig=px.bar(df_top10, x='Player', y='PTS', title='PPG by Player', width=1200, height=800, color_discrete_sequence=['purple'])
fig.update_xaxes(title='player')
fig.update_yaxes(title='pts')
fig.show()
```

```python
#create a scaled dataframe

#establish your descriptive columns and your numeric columns
des_cols = ['Player','Pos','Tm','Rk','Age']

num_cols = df.columns.tolist()
for col in des_cols:
    num_cols.remove(col)

#create dataset version with only numeric columns
num_data = df.drop(des_cols,axis=1)

#scale the numeric data
scaler = MinMaxScaler()
num_df_scaled = scaler.fit_transform(num_data.to_numpy())
num_df_scaled = pd.DataFrame(num_df_scaled, columns=num_cols)

#add the descriptive columns to the scaled tata set using merge
df_scaled = pd.merge(df[des_cols], num_df_scaled, left_index=True, right_index=True)

print("Scaled Dataset Using MinMaxScaler")
df_scaled.head()
```

```python
#sub plots using bar charts

mainstats = ['TRB','AST','STL','BLK','TOV','PTS']
top_players = df_scaled['Player'][0:10]


i=0
for player in top_players:  
    x_data = mainstats
    y_data = df_scaled.loc[df_scaled['Player']==player,mainstats].to_numpy()[0]
    fig = px.bar(x=x_data, y=y_data, title=player, width=1200, height=400, color_discrete_sequence=['light blue'])
    fig.update_xaxes(title='', visible=True, showticklabels=True)
    fig.update_yaxes(title='', visible=True, showticklabels=True)
    fig.show()
    i=i+1
    
    
    
```

```python
pts_age_df = pd.pivot_table(df[['Age','PTS']],index= 'Age',values='PTS',aggfunc=np.mean).reset_index()
fig = px.scatter(pts_age_df, x='Age', y='PTS', title='Points by Age', width=1200, height=600)
fig.show()
```

```python
bf_line = np.polyfit(x=pts_age_df['Age'], y=pts_age_df['PTS'], deg=3)
a=bf_line[0]
b=bf_line[1]
c=bf_line[2]
d=bf_line[3]

x_vals=pts_age_df['Age']
y_vals=pts_age_df['PTS']
bf_values = a*x_vals**3 + b*x_vals**2 + c*x_vals + d

fig = px.line(x=x_vals, y=bf_values, title='average points by age fitted', width=1200, height=800)
fig.add_scatter(x=x_vals,y=y_vals, mode='markers')
fig.update_xaxes(title='age')
fig.update_yaxes(title='points')
fig.show()
```
