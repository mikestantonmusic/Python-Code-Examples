```python
import networkx as nx
import matplotlib
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
G.add_nodes_from(range(1,6))
G.add_edge(1,3)
G.add_edge(3,5)
pos=nx.circular_layout(G)
nx.draw(G,pos,with_labels=1)
```
