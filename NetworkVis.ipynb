```python
import networkx as nx
import matplotlib
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
G.add_nodes_from(range(1,8))
G.add_edge(1,3)
G.add_edge(3,5)
G.add_edge(5,7)
G.add_edge(1,4)
G.add_edge(4,7)
pos=nx.circular_layout(G)
nx.draw(G,pos,with_labels=1)
```

```python
edges_list = [(1,2,1),(1,3,2),(1,5,4),(2,3,3),(3,4,1),(4,5,2),(3,6,3),]

def numeric_network(edges):
    
    '''
    this function draws a network using a list of edges
    the third digit of the tuple is the weight
    the nodes are arranged in numeric value
    '''
    
    G = nx.Graph()
    for edge in edges:
        G.add_edge(edge[0],edge[1], weight = edge[2])

    pos = nx.circular_layout(G)
    pos2 = pos.copy()
    dict1 = dict(nx.degree(G))
    sorted_keys = sorted(dict1, key=dict1.get)

    i=1
    for w in pos.keys():
        pos[i] = pos2[w]
        i=i+1

    options = {"edge_color":"orange",}

    nx.draw_networkx(G, pos, **options)
    edge_labels = nx.get_edge_attributes(G, "weight")
    nx.draw_networkx_edge_labels(G, pos, edge_labels, horizontalalignment='right')
    ax = plt.gca()
    ax.margins(0.20)
    plt.axis("off")
    plt.show()
    
numeric_network(edges=edges_list)
```

```python
edges_list = [(1,2),(1,3),(1,5),(2,3),(3,4),(4,5),(3,6),]

def arranged_network(edges):
    
    '''
    this function draws a network using a list of edges
    the nodes are arranged in order of degree
    so the nodes with the most edges are next to each other
    then the nodes with the second most edges, and so on
    '''
    
    G = nx.Graph()
    for edge in edges:
        G.add_edge(edge[0],edge[1])

    pos = nx.circular_layout(G)
    pos2 = pos.copy()
    dict1 = dict(nx.degree(G))
    sorted_keys = sorted(dict1, key=dict1.get)
    for i, w in zip(sorted_keys, pos.keys()):
        pos[i] = pos2[w]

    options = {"edge_color":"orange",}

    nx.draw_networkx(G, pos, **options)
    ax = plt.gca()
    ax.margins(0.20)
    plt.axis("off")
    plt.show()
    
arranged_network(edges=edges_list)
```
